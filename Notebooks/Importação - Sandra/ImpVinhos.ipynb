Importação de Vinhos

```python
import pandas as pd
```

```python
# URL do arquivo CSV no GitHub
urlVinhos = 'https://raw.githubusercontent.com/Data-Analitycs-Pos-Tech-Fiap/Vinifera/refs/heads/main/Dataset/Importacao/ImpVinhos.csv'
# Leitura do arquivo CSV
impVinho = pd.read_csv(urlVinhos, sep=';')

# Exibindo as primeiras linhas do DataFrame
print(impVinho.head())
```

```python
impVinho.head()
```

```python
impVinho.info()
```

```python
pd.options.display.float_format= '{:,.2f}'.format
```

```python
# Seleção de dados Quantidade (Kg)
impVinhoKg = impVinho[['País','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023']]
```

```python
impVinhoKg.head()
```

```python
# Seleção de dados Valor (US$)
impVinhoDolar = impVinho[['País','2009.1','2010.1','2011.1','2012.1','2013.1','2014.1','2015.1','2016.1','2017.1','2018.1','2019.1','2020.1','2021.1','2022.1','2023.1']]
```

```python
impVinhoDolar.head()
```

```python
# Renomeando as colunas Valor (US$)
novaColuna = {col: col.split('.')[0] for col in impVinhoDolar.columns[1:]}
impVinhoDolar = impVinhoDolar.rename(columns=novaColuna)
```

```python
impVinhoDolar.head()
```

```python
impVinhoKg.describe()
```

```python
impVinhoDolar.describe()
```

```python
# Transformando uma coluna em índice
dados_usaveisKg = impVinhoKg.set_index("País")
dados_usaveisKg.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

axis = impVinhoKg.plot(x="País", y="2009", kind="bar", figsize=(20,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.xticks(rotation=45)
plt.title("Quantidade de Vinho Importado por País")
plt.xlabel("País")
plt.ylabel("Quantidade (Kg)")

# Não faz retornar nenhum comentário, apenas o gráfico na "tela"
plt.show()
```

```python
dados_usaveisKg.T.plot(figsize=(10,6))
plt.title("Volume em Litros por País", fontsize=18, fontweight='bold', color='blue', loc='center', pad=20)
# Colocando a legenda fora do gráfico
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
# Identificando os 7 países com os maiores valores médios (por linha)
paises_topUs = impVinhoDolar.mean(numeric_only=True, axis=1).sort_values(ascending=False).head(6).index.tolist()

paises_topUs
```

```python
# Filtrando os dados para os 7 países
dados_topUs = impVinhoDolar.loc[paises_topUs]
dados_topUs
```

```python
# Transformando uma coluna em índice
dados_usaveisUs = dados_topUs.set_index("País")
dados_usaveisUs.head()
```

```python
dados_usaveisUs = dados_usaveisUs / 1_000_000
```

```python
# Plotando o gráfico do Valor em Dólar da Importação
dados_usaveisUs.T.plot(figsize=(10,6))
plt.title("Importação de Vinho: Valor em US$ por País (Top 6)", fontsize=18, fontweight='bold', color='blue', loc='center', pad=20)
plt.legend(loc='upper left')
plt.ylabel('Em milhões')
plt.show()
```

```python
paises_topKg = impVinhoKg.mean(numeric_only=True, axis=1).sort_values(ascending=False).head(6).index.tolist()
paises_topKg
```

```python
dados_topKg = impVinhoKg.loc[paises_topKg]
dados_topKg
```

```python
dados_usaveisKg = dados_topKg.set_index("País")
dados_usaveisKg
```

```python
dados_usaveisKg = dados_usaveisKg / 1_000_000
```

```python
# Calculando o total por linha e adicionando como nova coluna
indexPaisKg = impVinhoKg.set_index('País')
indexPaisKg['TotalQuant'] = indexPaisKg.sum(axis=1)
indexPaisKg.head()
```

```python
# Calculando o total por linha e adicionando como nova coluna
indexPaisDolar = impVinhoDolar.set_index('País')
indexPaisDolar['TotalDolar'] = indexPaisDolar.sum(axis=1)
indexPaisDolar.head()
```

```python
# Realizando o join
resultado = indexPaisDolar.join(indexPaisKg['TotalQuant'])
resultado.head()
```

```python
# Selecionando apenas as colunas 'TotalDolar' e 'TotalQuant' de 2009 a 2023
df_Totais = resultado[['TotalDolar', 'TotalQuant']]
df_Totais.head()
```

```python
# Ordenando por 'TotalDolar' e 'TotalQuant' em ordem descendente (2009 a 2023)
df_ordenado = df_Totais.sort_values(by=['TotalDolar', 'TotalQuant'], ascending=False).head(10)
df_ordenado
```

```python
df_ordem = df_ordenado.sort_values(by=['TotalDolar', 'TotalQuant'],ascending=True)
df_ordem
```

```python
df_ordem = df_ordem / 1_000_000
```

```python
import matplotlib.pyplot as plt

# Criando o gráfico de barras horizontais
ax = df_ordem.plot(kind='barh', figsize=(13, 7), y=['TotalDolar', 'TotalQuant'], label=['Valor (US$)', 'Volume (Litros)'])

# Adicionando os valores nas barras
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate('{:,.0f}M'.format(width), (x + width + 0.02, y + height/2), ha='left')
#plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.0f}K'.format(x/1000)))
plt.xlabel('Valor Gasto em Milhões de Reais')
plt.ylabel('10 Maiores Importadores ao Brasil')
plt.title('Importação de Vinhos entre 2009 a 2023', fontsize=16, fontweight='bold', color='blue', loc='center', pad=20)
plt.legend(loc='right', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
# Setar o índice como 'País'
TAnoDolar = impVinhoDolar.set_index('País')

# Adicionar uma nova linha com os totais por coluna
TAnoDolar.loc['TotalAnoValor'] = TAnoDolar.sum(axis=0)
TAnoDolar.tail()
```

```python
# Removendo a coluna 'Total'
#TAnoDolarL = TAnoDolar.drop('TotalDolar', axis=1)
TAnoDolarL = TAnoDolar
TAnoDolarL.tail()
```

```python
# Selecionando a linha 'TotalAnoDolar'
ImpTotalDolarL = TAnoDolarL.loc['TotalAnoValor']
ImpTotalDolarL
```

```python
# Setar o índice como 'País'
TAnoKg = impVinhoKg.set_index('País')

# Adicionar uma nova linha com os totais por coluna
TAnoKg.loc['TotalAnoKg'] = TAnoKg.sum(axis=0)
TAnoKg.tail()
```

```python
# Selecionando a linha 'TotalAnoKgL'
ImpTotalKgL = TAnoKg.loc['TotalAnoKg']
ImpTotalKgL
```

```python
# Realizando o merge com base nos índices
df_merged = pd.merge(ImpTotalDolarL, ImpTotalKgL, left_index=True, right_index=True)
df_merged.head()
```

```python
df_merged = df_merged / 1_000_000
```

```python
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

# Criando o gráfico de linhas
# Convert the index to numeric if it's not already
# df_merged.index = pd.to_numeric(df_merged.index, errors='coerce') # Convert to numeric, handle errors

axis = df_merged.plot(kind='line', marker='o', linewidth=2, figsize=(10, 6), y=['TotalAnoValor', 'TotalAnoKg'], label=['Valor (US$)', 'Volume (Litros)'])
# axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
# Personalizando o gráfico (opcional)
plt.title('Evolução do Total em Dólares e Volume por Ano', fontsize=16, fontweight='bold', color='blue', loc='center', pad=20)
# plt.xlabel('Ano', fontsize=12)
plt.ylabel('Em Milhões', fontsize=12)
#plt.grid(True)
# Adicionar rótulos nos pontos

# Adicionar rótulos nos pontos, ajustando a posição para evitar sobreposição
for index, row in df_merged.iterrows():
    plt.annotate(f"{row['TotalAnoValor']/1:.0f}", (index, row['TotalAnoValor']), xytext=(0, -10), textcoords='offset points')
    plt.annotate(f"{row['TotalAnoKg']/1:.0f}", (index, row['TotalAnoKg']), xytext=(0, -10), textcoords='offset points')


plt.legend()

# Formatando o eixo y para mostrar os valores em milhares
#plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:.0f}K'.format(x/1000)))

# Exibindo o gráfico
plt.show()
```
